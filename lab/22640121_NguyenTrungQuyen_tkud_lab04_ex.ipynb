# KHÁM PHÁ DỮ LIỆU

Dữ liệu chứa trong file `tips.csv` chứa thông tin tiền tip của khách hàng tại một nhà hàng. Dữ liệu có các thông tin sau:

- `total_bill`: tổng hoá đơn (USD)
- `tip`: tiền tip/ tiền boa (USD)
- `sex`: giới tính người thanh toán hóa đơn
- `smoker`: có hút thuốc hay không
- `day`: ngày trong tuần
- `time`: thời gian dùng bữa(buổi sáng, buổi trưa, buổi tối)
- `size`: số lượng người tham gia

Bạn hãy thực hiện thống kê mô tả để tìm hiểu về hành vi tip của khách hàng tại nhà hàng

## HIỂU DỮ LIỆU

1. Đọc dữ liệu
2. Xoá cột `Unnamed: 0`
3. Cho biết dữ liệu có bao nhiêu dòng, bao nhiêu thuộc tính.
4. Trong các thuộc tính, thuộc tính nào là thuộc tính phân loại, thuộc tính nào là thuộc tính định lượng
5. Có bao nhiêu nam, bao nhiêu nữ. Tỷ lệ bao nhiêu?
6. Có bao nhiêu người hút thuốc. Tỷ lệ bao nhiêu?
7. Ngày nào trong tuần khách hàng hay đến dùng bữa nhất. Tỷ lệ bao nhiêu?
8. Nhà hàng phục vụ bữa ăn nào nhiều nhất (Breakfast, Lunch, Dinner)?
9. Nhóm khách hàng dùng bữa tại nhà hàng nhiều nhất là bao nhiêu người?
10. Trung bình trong một bữa ăn nhà hàng phục vụ mấy khách?

```python
#1. Đọc dữ liệu
import pandas as pd
data = pd.read_csv("tips.csv")
data.head()
```

```python
#2. Xoá cột `Unnamed: 0`
data = data.drop(columns = ['Unnamed: 0'])
data.head()
```

```python
#3. Cho biết dữ liệu có bao nhiêu dòng, bao nhiêu thuộc tính.
data.shape
```

```python
#4. Trong các thuộc tính, thuộc tính nào là thuộc tính phân loại, thuộc tính nào là thuộc tính định lượng
# thuộc tính phân loại:Sex, smoker, day, time
data.dtypes[data.dtypes == 'object']
# thuộc tính định lượng: total_bill, tip, size
data.dtypes[data.dtypes != 'object']
```

```python
#5. Có bao nhiêu nam, bao nhiêu nữ. Tỷ lệ bao nhiêu?
sex = data['sex']
# Số lượng nam
print('số lượng nam:', (sex == 'Male').sum())
# Số lượng nữ   
print('số lượng nữ:', (sex == 'Female').sum())
# Tỷ lệ nam 
print('tỷ lệ nam:',(sex == 'Male').sum()/ sex.size)
# Tỷ lệ nữ
print('tỷ lệ nữ:', (sex == 'Female').sum() / sex.size)
```

```python
#6. Có bao nhiêu người hút thuốc. Tỷ lệ bao nhiêu?
smoker = data['smoker']
# Số lượng người hút thuốc
print('số lượng người hút thuốc:', (smoker == 'Yes').sum())
# tỷ lệ người hút thuốc
print('tỷ lệ người hút thuốc:', (smoker == 'Yes').sum() / smoker.size)
```

```python
#7. Ngày nào trong tuần khách hàng hay đến dùng bữa nhất. Tỷ lệ bao nhiêu?
day = data['day']
# Số lần khách hàng đến dùng bữa theo ngày
day.value_counts() 
# Ngày khách hàng hay đến dùng bữa nhất
print('Ngày khách hàng hay đến dùng bữa nhất:', day.value_counts().idxmax())
# Tỷ lệ khách hàng đến dùng bữa nhất
print('Tỷ lệ khách hàng đến dùng bữa nhất:', day.value_counts().max() / day.size)
```

```python
#8. Nhà hàng phục vụ bữa ăn nào nhiều nhất (Breakfast, Lunch, Dinner)?
time = data['time']
# Số lần khách hàng đến dùng bữa theo thời gian
time.value_counts()
# Bữa ăn nào nhiều nhất
print('Bữa ăn nào nhiều nhất:', time.value_counts().idxmax())
```

```python
#9. Nhóm khách hàng dùng bữa tại nhà hàng nhiều nhất là bao nhiêu người?
size = data['size']
# Số lần khách hàng dùng bữa theo số người
size.value_counts()
# Nhóm khách hàng dùng bữa nhiều nhất
print('Nhóm khách hàng dùng bữa nhiều nhất:', size.value_counts().idxmax())
```

```python
#10. Trung bình trong một bữa ăn nhà hàng phục vụ mấy khách?
print('Trung bình trong một bữa ăn nhà hàng phục vụ:', size.mean())
```

## YẾU TỐ ẢNH HƯỞNG ĐẾN TIỀN TIP

Trong phần này, chúng ta thử tìm hiểu về tiền tip tại nhà hàng và các yếu tố có thể ảnh hưởng đến nó

1. Tỷ lệ nhận được tiền tip của nhà hàng là bao nhiêu?
2. Tỷ lệ tiền tip trên tổng hóa đơn là bao nhiêu?
3. Giữa nam và nữ ai tip nhiều tiền hơn, ai tip thường xuyên hơn?
4. Giữa người hút thuốc và người không hút thuốc ai tip nhiều tiền hơn, ai tip thường xuyên hơn?
5. Thống kê tổng số tiền tip nhận được theo ngày.

```python
#1. Tỷ lệ nhận được tiền tip của nhà hàng là bao nhiêu?
tip = data['tip']
total_bill = data['total_bill']
# Tỷ lệ nhận được tiền tip
print('Tỷ lệ nhận được tiền tip:', tip.sum() / total_bill.sum())
```

```python
#2. Tỷ lệ tiền tip trên tổng hóa đơn là bao nhiêu?
# Tỷ lệ tiền tip trên tổng hóa đơn
print('Tỷ lệ tiền tip trên tổng hóa đơn:', tip.sum() / total_bill.sum())
```

```python
#3. Giữa nam và nữ ai tip nhiều tiền hơn, ai tip thường xuyên hơn?
# Tính tổng tiền tip của nam và nữ
sex_tip = data.groupby('sex')['tip'].sum()
print('Tổng tiền tip của nam:', sex_tip.get('Male'))
print('Tổng tiền tip của nữ:', sex_tip.get('Female'))
if(sex_tip.get('Male') > sex_tip.get('Female')):
    print('Nam tip nhiều hơn')
else:
    print('Nữ tip nhiều hơn')


```

```python
#4. Giữa người hút thuốc và người không hút thuốc ai tip nhiều tiền hơn, ai tip thường xuyên hơn?
# Tính tổng tiền tip của người hút thuốc và người không hút thuốc
smoker_tip = data.groupby('smoker')['tip'].sum()
print('Tổng tiền tip của người hút thuốc:', smoker_tip.get('Yes'))
print('Tổng tiền tip của người không hút thuốc:', smoker_tip.get('No'))
if(smoker_tip.get('Yes') > smoker_tip.get('No')):
    print('Người hút thuốc tip nhiều hơn')
else:
    print('Người không hút thuốc tip nhiều hơn')
```

```python
#5. Thống kê tổng số tiền tip nhận được theo ngày.
day_tip = data.groupby('day')['tip'].sum()
print('Tổng số tiền tip nhận được theo ngày:', day_tip)
```

## Với Matplotlib

Bạn hãy trực quan với thư viện matplotlib

```python
import matplotlib.pyplot as plt
#1. Vẽ đồ thị thể hiện tổng tiền tip theo ngày
day_tip.plot(kind = 'bar')
plt.xlabel('Day')
plt.ylabel('Total tip')
plt.title('Total tip by day')
plt.show()
```

```python
#2. Vẽ đồ thị thể hiện tổng tiền tip theo bữa ăn (time)
time_tip = data.groupby('time')['tip'].sum()
time_tip.plot(kind = 'bar')
plt.xlabel('Time')
plt.ylabel('Total tip')
plt.title('Total tip by time')
plt.show()
```

```python
#3. Vẽ đồ thị histogram của tổng hoá đơn
total_bill = data['total_bill']
total_bill.plot(kind = 'hist')
plt.xlabel('Total bill')
plt.ylabel('Frequency')
plt.title('Histogram of total bill')
plt.show()
```

```python
#4. Vẽ đồ thị boxplot của tổng hoá đơn
total_bill.plot(kind = 'box')
plt.ylabel('Total bill')
plt.title('Boxplot of total bill')
plt.show()
```

```python
#5. Vẽ đồ thị để kiểm tra sự liên quan giữa tổng hoá đơn và tiền tip.
plt.scatter(total_bill, tip)
plt.xlabel('Total bill')
plt.ylabel('Tip')
plt.title('Total bill vs Tip')
plt.show()
```

```python
#6. So sánh số tiền mà khách hàng đã chi giữa các ngày trong tuần.
total_bill_day = data.groupby('day')['total_bill'].sum()
total_bill_day.plot(kind = 'bar')
plt.xlabel('Day')
plt.ylabel('Total bill')
plt.title('Total bill by day')
plt.show()
```

## Với Seaborn
