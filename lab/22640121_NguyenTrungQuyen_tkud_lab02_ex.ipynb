## NUMPY

Bạn hãy phát sinh ngẫu nhiên mảng ndarray có tên `arr` gồm 100 phần tử số nguyên có giá trị từ 0..1000. Thực hiện các thao tác sau:

### 1. Hiển thị các phần tử thỏa yêu cầu dưới đây

- 10 phần tử đầu tiên
- 5 phần tử cuối cùng
- Các phần tử thỏa mãn điều kiện chia hết cho 5 và lớn hơn 100

```python

import numpy as np

arr = np.random.randint(0,1000,100)
print(arr)
print('10 phần tử đầu tiên:',arr[:10])
print('5 phần tử cuối cùng',arr[-5:])
print('các phần tử chia hết cho 5 và lớn hơn 100',arr[(arr % 5 == 0) & (arr > 100)])

```

### 2. Cho biết mảng trên có bao nhiêu phần tử có giá trị nhỏ hơn 100. Hiển thị các phần tử đó

```python
print('mảng arr có',len(arr[arr < 100]),'phần tử nhỏ hơn 100')
print('các phần tử nhỏ hơn 100:',arr[arr < 100])
```

### 3. Thực hiện thống kê:

- Tính tổng mảng
- Tìm giá trị trung bình
- Tìm giá trị trung vị
- Tìm Q1, Q2, Q3
- Tìm phân vị 20 (percentile 20)

```python
print('tổng mảng:',np.sum(arr))
print('giá trị trung bình:',np.mean(arr))
print('giá trị trung vị:',np.median(arr))
print('Q1:',np.percentile(arr,25))
print('Q2:',np.percentile(arr,50))
print('Q3:',np.percentile(arr,75))
print('phân vị 20:',np.percentile(arr,20))
```

### 4. Thực hiện việc chọn ngẫu nhiên (không thay thế) 10 phần tử từ mảng trên 5 lần. Hiển thị kết quả lựa chọn mỗi lần

```python
for i in range(5):
    print('lần',i+1,':',np.random.choice(arr,10,replace=False))
```

### 5. Tìm phần tử xuất hiện nhiều nhất trong mảng

```python
values, counts = np.unique(arr, return_counts=True)
print('phần tử xuất hiện nhiều nhất trong mảng là:',values[np.argmax(counts)])
```

### Câu 4: Định dạng đường thẳng

Phát sinh x là các điểm dữ liệu từ $[0, \pi]$. Bạn hãy vẽ đồ thị $y = sin(x)$ và $y = cos(x)$ như dưới đây.
Gợi ý: sử dụng hằng số pi và hàm sau:
- `np.pi` : số $\pi$
- `np.linspace()`
- `np.sin()`
- `np.cos()`

```python
# This cell contains output. Don't remove!
```

```python
# YOUR CODE HERE
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 2 * np.pi, 100)  

y1 = np.sin(x)
y2 = np.cos(x)

plt.plot(x, y1, label='sin(x)', color='blue') 
plt.plot(x, y2, label='cos(x)', color='orange', linestyle='--') 



plt.title('Đồ thị hàm sin(x) và cos(x)')

plt.legend()

plt.grid()

plt.show()
```

### Câu 10: Kết hợp trên cùng một đồ thị

Sử dụng hàm `np.random.normal(size=10000)` để phát sinh 10000 số ngẫu nghiên thuộc phân phối chuẩn. Vẽ hình thể hiện phân phối như sau:

Hướng dẫn: thêm tham số `vert=False` trong hàm vẽ `boxplot()`

```python
# This cell contains output. Don't remove!
```

```python
import numpy as np
import matplotlib.pyplot as plt

data = np.random.normal(size=10000)

fig, axs = plt.subplots(2, 1, figsize=(8, 6))
fig.suptitle('đồ thị hàm sin(x) và cos(x)')

# Biểu đồ histogram
axs[0].hist(data, bins=100, density=True, color='purple', alpha=0.7)
axs[0].set_title('Histogram')
axs[0].set_ylabel('Mật độ')

# Biểu đồ boxplot
axs[1].boxplot(data, vert=False)
axs[1].set_title('Boxplot')
axs[1].set_xlabel('Giá trị')

# Hiển thị biểu đồ

plt.show()
```
