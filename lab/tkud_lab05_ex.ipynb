```python
# Xac suat de so cham xuat hien la 4 theo phan phoi nhi thuc P(k)= kCn * p^k * ( 1 - p) ^ (n - k) voi P(4)

# Mô phỏng tung xúc sắc 5000 lần (mỗi lần giá trị từ 1 đến 6)
n = 5000
rolls = np.random.randint(1, 7, size=n)
print(rolls)
# Xs de so cham la 4 thi nhu la lay 1 mac trong 6 mac cua xuc sac, voi lay 1 mac la 1/6 thi lay mac 4 la 1/6
prob_4 = np.sum(rolls == 4) / n
print(f"Xác suất xuất hiện số 4: {prob_4:.4f} gan bang 1/6")

# Xs de so cham lon hon hoac bang 4 thi ta dung xs voi 1 mac la 1/6 thi lay mac 4 la 1/6, 5 la 1/6, 6 la 1/6
prob_4_5_6 = np.sum(rolls >= 4) / n
print(f"Xác suất xuất hiện số lớn hơn hoặc bằng 4: {prob_4:.4f} gan bang 1/2")

# voi prob_4_5_6 la xs so cham lon hon hoac bang 4 thi xs mat 6 cham la ty le la 1/3 
prob_6 = np.sum(rolls == 6) / n
print(f"Xác suất xuất hiện số 6: {prob_6:.4f} gan bang 1/3 ")
```

**Bài 10**

Viết một chương trình để mô phỏng việc tung đồng xu và đánh giá phương sai của giá trị mô phỏng của biến ngẫu nhiên X (tung được mặt `Head`) đối với các số lần tung khác nhau. Giả sử  đồng xu phải cân bằng.

Đối với mỗi N từ 10; 40; 90; 160; 250; 490; 640; 810; 1000, hãy ước tính giá trị của X bằng cách mô phỏng số lần tung đó.

Bạn nên chạy mỗi mô phỏng 100 lần và sử dụng tập hợp các ước tính để đánh giá phương sai của ước tính x của bạn. Vẽ biểu đồ phương sai này so với N, bạn có nhận xét gì không?

```python
def tinh_Phuong_Sai(N_values, n=100):
    v = []
    
    for N in N_values:
        e = []
        for _ in range(n):
            flips = np.random.choice([0, 1], size=N, p=[0.5, 0.5])  # 0: Tail, 1: Head
            e.append(np.mean(flips))
        
        variance = np.var(e)
        v.append(variance)
    
    return N_values, v

def plot_variance(N_values, v):
    """Vẽ biểu đồ phương sai theo số lần tung N."""
    plt.figure(figsize=(8, 5))
    plt.plot(N_values, v, marker='o', linestyle='-', color='b', label="Phương sai thực nghiệm")
    plt.xlabel("Số lần tung N")
    plt.ylabel("Phương sai của X")
    plt.title("Biểu đồ phương sai của X theo số lần tung N")
    plt.grid(True, linestyle="--", alpha=0.7)
    plt.legend()
    plt.show()

# Chạy mô phỏng với danh sách N_values cụ thể
N_values = [10, 40, 90, 160, 250, 490, 640, 810, 1000]
n = 100

N_vals, v = tinh_Phuong_Sai(N_values, n)
plot_variance(N_vals, v)
print("Nhan xet thi thay phuong sai cua X giam khi tang so lan tung N"
"\n co nghi la cang nhieu gia tri thi do phan tan cua cac gia tri se giam dan")
```
