**Bài 05:** Tạo 5000 số ngẫu nhiên có phân phối nhị thức với n=50, p=0.7 Bạn Hãy

a. Tìm các giá trị thống kê: min, max, Q1, Q2, Q3

b. Vẽ đồ thị boxplot

c. Vẽ các đồ thị: tần số, tần suất

```python
#a. Tìm các giá trị thống kê: min, max, Q1, Q2, Q3
n = 50
p = 0.7
size = 5000
data = np.random.binomial(n, p, size)

# Tính các giá trị thống kê
min_val = np.min(data)
max_val = np.max(data)
q1 = np.percentile(data, 25)
q2 = np.median(data)
q3 = np.percentile(data, 75)

print("Các giá trị thống kê:")
print("Min: ",min_val)
print("Max: ",max_val)
print("Q1 : ", q1)
print("Q2 : ", q2)
print("Q3 : ", q3)
```

```python
#. Vẽ đồ thị boxplot
plt.figure(figsize=(8, 6))
plt.boxplot(data)
plt.title('Boxplot của phân phối nhị thức')
plt.ylabel('Giá trị')
plt.show()
```

```python
#c. Vẽ các đồ thị: tần số, tần suất
# Tính tần số
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.hist(data, bins=range(min_val, max_val + 2), edgecolor='black', alpha=0.7)
plt.title('Tần số của các giá trị')
plt.xlabel('Giá trị')
plt.ylabel('Tần số')

# Tính tần suất
plt.subplot(1, 2, 2)
plt.hist(data, bins=range(min_val, max_val + 2), density=True, edgecolor='black', alpha=0.7)
plt.title('Tần suất của các giá trị')
plt.xlabel('Giá trị')
plt.ylabel('Tần suất')

plt.tight_layout()
plt.show()
```

Giả sử điểm thi TOEIC là một biến ngẫu nhiên có phân phối chuẩn với kỳ vọng là 500, độ lệch chuẩn là 8. Bạn hãy:

a. Mô phỏng bằng cách phát sinh ngẫu nhiên điểm thi TOIEC của một nhóm gồm 50 người.

b. Tính điểm trung bình và độ lệch chuẩn về điểm của nhóm.

c. Tìm miền giá trị, và miền phân vị (IQR) của nhóm

d. Cho biết tỷ lệ đạt trên 450 điểm của nhóm. 

e. Vẽ đồ thị histogram của nhóm và so sánh với đồ thị hàm mật độ của phân phối chuẩn với kỳ vọng là 500 và độ lệch chuẩn là 8.

f. Thay vì mô phỏng 50 người, hãy mô phỏng nhóm với 100 người thi. Bạn có nhận xét gì?

```python
#Giả sử điểm thi TOEIC là một biến ngẫu nhiên có phân phối chuẩn với kỳ vọng là 500, độ lệch chuẩn là 8. Bạn hãy:

mu_toeic = 500
sigma_toeic = 8
#a. Mô phỏng bằng cách phát sinh ngẫu nhiên điểm thi TOIEC của một nhóm gồm 50 người.
toeic_scores_50 = np.random.normal(mu_toeic, sigma_toeic, 50)
```

```python
#b. Tính điểm trung bình và độ lệch chuẩn về điểm của nhóm.
mean_50 = np.mean(toeic_scores_50)
std_50 = np.std(toeic_scores_50)
print("Điểm trung bình của nhóm 50 người:", mean_50)
print("Độ lệch chuẩn về điểm của nhóm 50 người:", std_50)
```

**Bài 10**

Viết một chương trình để mô phỏng việc tung đồng xu và đánh giá phương sai của giá trị mô phỏng của biến ngẫu nhiên X (tung được mặt `Head`) đối với các số lần tung khác nhau. Giả sử  đồng xu phải cân bằng. 

Đối với mỗi N từ 10; 40; 90; 160; 250; 490; 640; 810; 1000, hãy ước tính giá trị của X bằng cách mô phỏng số lần tung đó. 

Bạn nên chạy mỗi mô phỏng 100 lần và sử dụng tập hợp các ước tính để đánh giá phương sai của ước tính x của bạn. Vẽ biểu đồ phương sai này so với N, bạn có nhận xét gì không?

```python
N_values = [10, 40, 90, 160, 250, 490, 640, 810, 1000]
num_simulations = 100 

variances = []  

for N in N_values:
    estimates = []  
    for _ in range(num_simulations):
        tosses = np.random.choice([0, 1], size=N, p=[0.5, 0.5]) 
        estimate_X = np.mean(tosses)  
        estimates.append(estimate_X)
    variance_X = np.var(estimates)  
    variances.append(variance_X)

# Vẽ biểu đồ phương sai theo N
plt.figure(figsize=(8, 5))
plt.plot(N_values, variances, marker='o', linestyle='-', color='b', label='Phương sai của X')
plt.xlabel('Số lần tung (N)')
plt.ylabel('Phương sai')
plt.title('Biểu đồ phương sai của biến ngẫu nhiên X theo N')
plt.legend()
plt.grid()
plt.show()

# Nhận xét:
# Khi N tăng, phương sai giảm dần, chứng tỏ rằng việc tăng số lần tung giúp ước tính ổn định hơn.
```
