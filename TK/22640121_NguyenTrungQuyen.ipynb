```python
import pandas as pd
```

```python
df = pd.read_csv('thoi_tiet_khong_khi_vn_2023.csv')
```

```python
#1.	(0.75đ) Hiển thị 10 dòng đầu tiên và 10 dòng cuối cùng của bảng dữ liệu.
print(pd.concat([df.head(10), df.tail(10)]))
```

```python
#2.	(0.75đ) Cho biết bảng dữ liệu có bao nhiêu dòng và bao nhiêu cột?
df.shape
```

```python
#3.	(0.75đ) Cho biết thông tin về các cột dữ liệu và kiểu của từng cột trong bảng.
df.info()
```

```python
#4.	(0.75đ) Cho biết số lượng bản ghi bị thiếu giá trị cho các cột: Nhiệt_độ, Lượng_mưa và AQI.
print("số lượng bản ghi bị thiếu của nhiệt độ",df["Nhiệt_độ"].isnull().sum())
print("số lượng bản ghi bị thiếu của lượng mưa",df["Lượng_mưa"].isnull().sum())
print("số lượng bản ghi bị thiếu của AQI",df["AQI"].isnull().sum())
```

```python
#1.	(0.75đ) Trung bình (mean), trung vị (median) và mode của nhiệt độ.
print('trung bình của nhiệt độ:',df["Nhiệt_độ"].mean())
print('trung vị của nhiệt độ:',df["Nhiệt_độ"].median())
print('mode nhiệt của độ:',df["Nhiệt_độ"].mode())
```

```python
#2.	(0.75đ) Độ lệch chuẩn (standard deviation) và phương sai (variance).
print ('Độ lệch chuẩn của nhiệt độ',df["Nhiệt_độ"].std())
print ('Phương sai của nhiệt độ',df["Nhiệt_độ"].var())
```

```python
#3.	(0.75đ) Nhiệt độ thấp nhất, cao nhất và biên độ nhiệt (range).
print ('Nhiệt độ thấp nhất',df["Nhiệt_độ"].min())
print ('Nhiệt độ cao nhất',df["Nhiệt_độ"].max())
print ('Biên độ nhất',df["Nhiệt_độ"].max()-df["Nhiệt_độ"].min())
```

```python
#1.	(0.75đ) Đếm số ngày có thời tiết "Nắng" và số ngày có thời tiết "Mưa to" hoặc "Dông".
thoi_tiet = df['Thời_tiết']
print('số ngày có thời tiết "Nắng":',(thoi_tiet=="Nắng").sum())
print('số ngày có thời tiết "Mưa to":',(thoi_tiet=="Mưa to").sum())
print('số ngày có thời tiết "Dông":',(thoi_tiet=="Dông").sum())
```

```python
#2.	(0.75đ) Vẽ đồ thị Pie để so sánh phân bố các loại thời tiết trong năm.
import matplotlib.pyplot as plt
y = df['Thời_tiết']
plt.pie(y)
plt.show()
```

```python
#1.	(0.75đ) Liệt kê 20 ngày có chỉ số AQI cao nhất và thành phố tương ứng.
df.sort_values(by='AQI', ascending=False).head(20)
```

```python
#2.	(0.75đ) Tính nhiệt độ trung bình theo tháng cho mỗi thành phố và hiển thị kết quả.
df['Ngày'] = pd.to_datetime(df['Ngày'], format='%Y-%m-%d')
df['Tháng'] = df['Ngày'].dt.month
df['Năm'] = df['Ngày'].dt.year

df.groupby(['Thành_phố', 'Tháng']).mean()['Nhiệt_độ']
grbyTheoTP = df.groupby(['Ngày','Thành_phố'])
print("nhiệt độ trung bình theo tháng cho mỗi thành phố",grbyTheoTP['Nhiệt_độ'].mean())
```

```python
#3.	(0.25đ) Cho biết thành phố có lượng mưa trung bình năm cao nhất. 
ThanhPho = df.groupby[('Thành_phố','Lượng_mưa'.mean())]
print("thành phố có lượng mưa trung bình năm cao nhất",ThanhPho)
```
