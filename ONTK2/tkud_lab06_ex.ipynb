### Bài 1: Tạo quần thể mô phỏng

Bạn hãy tạo một mẫu gồm 10000 phần tử mô phỏng chiều cao của nam thanh niên có giá trị từ 120cm - 200cm. Bạn hãy lưu kết quả vào biến **POP**.

Tính:
a. Chiều cao trung bình (kỳ vọng) của quần thể và độ lệch chuẩn về chiều cao của quần thể.

b. Tính tỷ lệ người cao trong quần thể, biết rằng thanh niên có chiều cao từ 180 trở lên được gọi là cao.

c. Vẽ histogram về chiều cao của quần thể. Theo bạn quần thể có phân phối chuẩn hay không?

d. Thử vẽ histogram và đồ thị hàm mật độ của phân phối chuẩn sử dụng tham số loc và scale bằng với kỳ vọng và độ lệch chuẩn của quần thể.

```python
# Import các thư viện thông dụng
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Import các distribution packages từ thư viện scipy
from scipy.stats import binom
from scipy.stats import poisson
from scipy.stats import norm
from scipy.stats import t
```

```python

#a. Chiều cao trung bình (kỳ vọng) của quần thể và độ lệch chuẩn về chiều cao của quần thể.
POP = np.random.uniform(120, 200, 10000)
mean_height = np.mean(POP)
std_dev_height = np.std(POP)

print("Chiều cao trung bình (kỳ vọng):", mean_height)
print("Độ lệch chuẩn:", std_dev_height)
```

```python
#b. Tính tỷ lệ người cao trong quần thể, biết rằng thanh niên có chiều cao từ 180 trở lên được gọi là cao.
# Tính kỳ vọng (mean) và độ lệch chuẩn (std) của quần thể
tall_rate = np.sum(POP >= 180) / len(POP)
print(f"Tỷ lệ người cao (>=180cm):" ,tall_rate*100)
```

```python
import matplotlib.pyplot as plt
#c. Vẽ histogram về chiều cao của quần thể. Theo bạn quần thể có phân phối chuẩn hay không?
plt.figure(figsize=(10, 6))
plt.hist(POP, bins=30, color='skyblue', edgecolor='black', alpha=0.7)
plt.title("Histogram về chiều cao của quần thể")
plt.xlabel("Chiều cao (cm)")
plt.ylabel("Tần suất")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

# Đánh giá:
# Histogram này cho thấy phân phối của quần thể. Nếu hình dạng của histogram gần giống hình chuông
# (đối xứng và tập trung quanh giá trị trung bình), thì quần thể có thể được coi là phân phối chuẩn.
```

```python
from scipy.stats import norm

# Td. Thử vẽ histogram và đồ thị hàm mật độ của phân phối chuẩn sử dụng tham số loc và scale bằng với kỳ vọng và độ lệch chuẩn của quần thể.
mean_height = np.mean(POP)
std_dev_height = np.std(POP)


plt.figure(figsize=(10, 6))
plt.hist(POP, bins=30, density=True, color='skyblue', edgecolor='black', alpha=0.7, label='Histogram của quần thể')


x = np.linspace(120, 200, 1000) 
pdf = norm.pdf(x, loc=mean_height, scale=std_dev_height)  
plt.plot(x, pdf, color='red', linewidth=2, label='Hàm mật độ của phân phối chuẩn')


plt.title("Histogram và hàm mật độ của phân phối chuẩn")
plt.xlabel("Chiều cao (cm)")
plt.ylabel("Mật độ xác suất")
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

### Bài 7:  Bài Tập Ước Lượng 

Khảo sát cân nặng (kg) của gà khi xuất chuồng, người ta cân một số con và có kết quả như sau:

> 2.1, 1.8, 2.0, 2.3, 1.7, 1.5, 2.0, 2.2, 1.8

Giả sử cân nặng của gà là biến ngẫu nhiên có phân phối chuẩn. Hãy lập khoảng tin cậy cho cân nặng trung bình của  khi xuất chuồng với độ tin cậy là 95%. Trong 2 trường hợp:

a, Biết $\sigma = 0.3$

b, Không biết $\sigma$


```python
import numpy as np
from scipy.stats import norm, t

# Dữ liệu mẫu
sample = np.array([2.1, 1.8, 2.0, 2.3, 1.7, 1.5, 2.0, 2.2, 1.8])
n = len(sample)
mean = np.mean(sample)
std_sample = np.std(sample, ddof=1)
```

```python

# a, Biết sigma
sigma = 0.3
z = norm.ppf(0.975)  # 95% CI → alpha/2 = 0.025
margin_error_z = z * (sigma / np.sqrt(n))
ci_z = (mean - margin_error_z, mean + margin_error_z)

print("Khoảng tin cậy 95% (biết σ=0.3):", {ci_z})
```

```python
#b ko bt sigma
t_value = t.ppf(0.975, df=n-1)
margin_error_t = t_value * (std_sample / np.sqrt(n))
ci_t = (mean - margin_error_t, mean + margin_error_t)

print("Khoảng tin cậy 95% (không biết σ):", {ci_t})
```

```python
df = pd.read_csv('SAT-lard.csv')
sat_scores = df['SAT Score']
df
```

```python
#a
n = len(sat_scores)
mean = np.mean(sat_scores)
std = np.std(sat_scores, ddof=1)

z_99 = norm.ppf(0.995) 
margin_error_99 = z_99 * (std / np.sqrt(n))
ci_99 = (mean - margin_error_99, mean + margin_error_99)

print(f"a. Khoảng tin cậy 99%: {mean:.2f} ± {margin_error_99:.2f}")
```

```python
#b 
z_95 = norm.ppf(0.975)  # z = 1.96 cho 95%
margin_error_95 = z_95 * (std / np.sqrt(n))
ci_95 = (mean - margin_error_95, mean + margin_error_95)

print(f"b. Khoảng tin cậy 95%: {mean:.2f} ± {margin_error_95:.2f}")
```

```python
mu = np.mean(sat_scores)
print(f"c. Trung bình SAT toàn trường: μ = {mu:.2f}")
```
