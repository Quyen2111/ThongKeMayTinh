```python
# Import necessary libraries
from scipy.stats import binom

# Define the parameters
n = 10  # Sample size
p = 2 / 100  # Probability of defective item (2 out of 100)

# Create a binomial distribution object
rv = binom(n, p)

# Calculate P(X = 0): Probability of 0 defective items
p_0 = rv.pmf(0)

# Calculate P(X = 1): Probability of 1 defective item
p_1 = rv.pmf(1)

# Total probability P(X <= 1) = P(X = 0) + P(X = 1)
p_at_most_1 = p_0 + p_1

# Print the results
print(f"Probability of 0 defective items: {p_0:.4f}")
print(f"Probability of 1 defective item: {p_1:.4f}")
print(f"Probability of at most 1 defective item: {p_at_most_1:.4f}")

# If the given answer suggests P(X >= 2), calculate that as well
print(binom.cdf(8 , n , p_at_most_1))
```

```python
from scipy.stats import poisson

# Tham số của phân phối Poisson
lam = 5  # lambda = 5

# Xác suất P(X = 1)
p_x_1 = poisson.pmf(1, lam)

# Xác suất P(X <= 1)
p_x_leq_1 = poisson.cdf(1, lam)

# Xác suất có điều kiện P(X >= 1 | X <= 1)
p_ketqua = p_x_1 / p_x_leq_1

print(f"Xác suất có điều kiện P(X >= 1 | X <= 1) = {p_ketqua:.4f}")
```

```python
import numpy as np
from scipy.stats import binom
# 2 cau thu
p1 = 0.6  # Xác suất trúng của cầu thủ 1
p2 = 0.7  # Xác suất trúng của cầu thủ 2

# Xác suất của X1 (ném 2 lần)
P_X1 = [binom.pmf(k, 2, p1) for k in range(3)]  # X1 có thể nhận giá trị 0, 1, 2
# Xác suất của X2 (ném 1 lần)
P_X2 = [binom.pmf(k, 1, p2) for k in range(2)]  # X2 có thể nhận giá trị 0, 1

# Khởi tạo mảng xác suất cho X (có thể nhận giá trị 0, 1, 2, 3)
P_X = np.zeros(4)

# Tính xác suất của X = X1 + X2
for i in range(3):  # X1 có thể là 0, 1, 2
    for j in range(2):  # X2 có thể là 0, 1
        P_X[i + j] += P_X1[i] * P_X2[j]

# In toàn bộ bảng phân phối xác suất của X
for i, p in enumerate(P_X):
    print(f"P(X = {i}) = {p:.3f}")
```

```python
from scipy.stats import hypergeom, binom, poisson
# thu pham
# Thông số bài toán
N = 25  # Tổng số sản phẩm trong hộp
K = 2   # Số thứ phẩm
n = 10  # Số sản phẩm được chọn
t = 8   # Số giờ làm việc

# Xác suất một lần kiểm tra đạt yêu cầu (≤1 thứ phẩm)
p_success = hypergeom.cdf(1, N, K, n)

# Xác suất ít nhất 7 lần đạt yêu cầu trong 8 giờ làm việc
p_binom = 1 - binom.cdf(6, t, p_success)
print(f"(a) Xác suất (nhị thức): {p_binom:.4f}")  # Kết quả: 0.6572
lambda_poisson = t * p_success
p_poisson = 1 - poisson.cdf(6, lambda_poisson)
print(f"(b) Xác suất (Poisson xấp xỉ): {p_poisson:.4f}")  # Kết quả: 0.6626
p_A = hypergeom.pmf(0, N, K, n)  # P(A): Không có thứ phẩm
p_B = p_success  # P(B): Đạt yêu cầu (≤1 thứ phẩm)

p_conditional = p_A / p_B
print(f"(c) Xác suất mẫu không có thứ phẩm khi đã đạt yêu cầu: {p_conditional:.4f}")  # Kết quả: 0.4118cv cv
```
