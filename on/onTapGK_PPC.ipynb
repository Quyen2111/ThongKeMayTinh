```python
import scipy.stats as stats

# Định nghĩa giá trị trung bình và độ lệch chuẩn tổng quát
def compute_probabilities(mu):
    sigma = 0.1 * mu

    # Phần (a): P(X > 1.15 * mu)
    L_a = 1.15 * mu
    prob_a = 1 - stats.norm.cdf(L_a, loc=mu, scale=sigma)

    return prob_a

# Phần (b): Với mu = 4, tìm L sao cho P(X ≤ L) = 0.1
mu_b = 4
sigma_b = 0.1 * mu_b
L_b = stats.norm.ppf(0.1, loc=mu_b, scale=sigma_b)

# Tính kết quả
prob_a_general = compute_probabilities(4)  # Cho mu = 4
L_b, prob_a_general
```

```python
import scipy.stats as stats

def probability_x_greater_than_80(mu, sigma):
    z_score = (79 - mu) / sigma
    return 1 - stats.norm.cdf(z_score)

# Thông số của X1 và X2
mu_1, sigma_1 = 140, 50
mu_2, sigma_2 = 200, 60

# Tính xác suất
p_x1 = probability_x_greater_than_80(mu_1, sigma_1)
p_x2 = probability_x_greater_than_80(mu_2, sigma_2)

print(f"Xác suất X1 ≥ 80: {p_x1:.4f}")
print(f"Xác suất X2 ≥ 80: {p_x2:.4f}")

# So sánh và đưa ra quyết định
if p_x1 > p_x2:
    print("Công ty nên chọn phương án 1.")
else:
    print("Công ty nên chọn phương án 2.")
```

```python
import scipy.stats as stats

# Thông số của phân phối chuẩn
mu = 175  # Trung bình
sigma = 4  # Độ lệch chuẩn

# (a) Xác suất chiều cao > 180 cm
p_a = 1 - stats.norm.cdf(180, mu, sigma)

# (b) Xác suất chiều cao nằm trong khoảng 166 cm đến 177 cm
p_b = stats.norm.cdf(177, mu, sigma) - stats.norm.cdf(166, mu, sigma)

# (c) Tìm h0 sao cho 33% người có chiều cao dưới mức h0
h0 = stats.norm.ppf(0.33, mu, sigma)

# (d) Giới hạn biến động của 90% chiều cao quanh giá trị trung bình
# (100 - 90 )/2 = 95
z_critical = stats.norm.ppf(0.95)  # Giá trị tới hạn cho mức 90% (symmetric)
range_d = z_critical * sigma

# In kết quả
print(f"(a) Xác suất chiều cao > 180 cm: {p_a:.3f}")
print(f"(b) Xác suất chiều cao trong khoảng 166 cm - 177 cm: {p_b:.3f}")
print(f"(c) h0: {h0:.2f} cm")
print(z_critical)
print(f"(d) Giới hạn biến động: {range_d:.1f} cm")
```
