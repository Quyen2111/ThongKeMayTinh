```python
xh = df[['vat_li', 'sinh_hoc', 'hoa_hoc']].isnull().sum(axis=1) < 3
tn = df[["dia_li", "lich_su", "gdcd"]].isnull().sum(axis=1) < 3
lanDau = (xh.sum() + tn.sum())
tuDo = df['sbd'].count() - lanDau
labels = ['Lan dau', 'Tu do']
import matplotlib.pyplot as plt
plt.pie([lanDau, tuDo], labels=labels)
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
df['TN'] =( df['vat_li'] + df['sinh_hoc'] + df['hoa_hoc'])/ 3
df['XH'] = ( df['lich_su'] + df['dia_li'] + df['gdcd']) / 3
mon_hoc = ['toan', 'ngu_van', 'ngoai_ngu', 'TN', 'XH']
for mon in mon_hoc:
    sns.histplot(df[mon], kde=True, label=mon, bins=50)
plt.xlabel("Điểm số")
plt.ylabel("Số lượng học sinh")
plt.title("Phổ điểm các môn học")
plt.legend()
plt.show()
```
