# PHÂN PHỐI XÁC SUẤT RỜI RẠC

### Biến ngẫu nhiên rời rạc
- Là biến ngẫu nhiên nhận các giá trị đếm được.
- Ký hiệu: X, Y, Z

### Hàm xác suất (pmf)
- `f(x) = P(X=x)` : Xác suất biến X nhận giá trị x.
- **Tính chất**:
  - `0 ≤ f(x) ≤ 1`
  - `∑f(x) = 1`

### Phân phối xác suất
- Mô tả tất cả giá trị x mà X có thể nhận và xác suất `P(X=x)` tương ứng.
- Biểu diễn: Bảng, đồ thị, hoặc công thức.

### Hàm phân phối tích lũy (cdf)
- `F(x) = P(X≤x)`: Tổng xác suất X nhận giá trị nhỏ hơn hoặc bằng x.

### Phân phối Nhị thức
- `X∼B(n,p)`: Số lần thành công trong n thử nghiệm độc lập, xác suất thành công p.
- Kỳ vọng: `μ = n * p`
- Phương sai: `σ^2 = n * p * q`
- Độ lệch chuẩn: `σ = sqrt(n * p * q)`

#### Điều kiện nhận biết
- **Số lần thử nghiệm cố định (n)**: Có một số lần thử cụ thể, hữu hạn và được xác định trước.
- **Hai kết quả khả dĩ**: Mỗi thử nghiệm chỉ có hai khả năng (thành công/thất bại, đúng/sai, trúng/miss, v.v.).
- **Xác suất cố định (p)**: Xác suất thành công trong mỗi lần thử là như nhau và không đổi.
- **Tính độc lập**: Các lần thử nghiệm không ảnh hưởng lẫn nhau.

### Phân phối Poisson
- `X∼P(λ)`: Số lần xuất hiện biến cố trong một khoảng, với λ là trung bình.
- Kỳ vọng: `μ = λ`
- Phương sai: `σ^2 = λ`
- Độ lệch chuẩn: `σ = sqrt(λ)`

#### Điều kiện nhận biết
1. **Không có số lần thử cố định**: Không có n cụ thể, biến cố xảy ra trong một khoảng liên tục (thời gian, không gian, v.v.).
2. **Biến cố hiếm**: Xác suất xảy ra biến cố trong một đơn vị khoảng rất nhỏ.
3. **Trung bình xuất hiện (λ)**: Số lần xuất hiện trung bình của biến cố trong khoảng đó là cố định.
4. **Tính độc lập**: Các biến cố xảy ra độc lập với nhau trong các khoảng khác nhau.
5. **Không giới hạn trên**: X có thể nhận giá trị từ 0 đến vô cực.

# PHÂN PHỐI CHUẨN

1. **Biến ngẫu nhiên liên tục**
   - Nhận giá trị trong khoảng liên tục, không đếm được.
   - `P(X=a)=0`

2. **Hàm mật độ xác suất (pdf)**
   - `P(a≤X<b)`: Xác suất X nằm trong khoảng [a,b)

3. **Phân phối chuẩn**
   - `X∼N(μ,σ^2)`: Đặc trưng bởi trung bình μ và phương sai σ^2

4. **Phân phối Z (Chuẩn chính tắc)**
   - `Z∼N(0,1)`: `μ=0`, `σ^2=1`

5. **Chuẩn hóa phân phối chuẩn**
   - `Z = (X − μ) / σ`: Biến đổi `X∼N(μ,σ^2)` về `Z∼N(0,1)`
