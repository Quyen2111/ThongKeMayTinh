```python
import numpy as np
import matplotlib.pyplot as plt

# Dữ liệu: Năm và doanh thu (giả định)
years = np.array([2018, 2019, 2020, 2021, 2022, 2023])
revenue = np.array([50, 55, 53, 60, 65, 70])

plt.plot(years, revenue, marker='o', linestyle='-', color='b', label="Doanh thu")
plt.xlabel("Năm")
plt.ylabel("Doanh thu (triệu USD)")
plt.title("Tăng trưởng doanh thu theo năm")
plt.legend()
plt.grid(True)
plt.show()
# Bieu do duong
# Dùng để hiển thị xu hướng dữ liệu theo thời gian hoặc mối quan hệ giữa hai biến.
```

```python
import numpy as np
import matplotlib.pyplot as plt
data3 = np.genfromtxt(r"D:\Quan_Ly_Mon_Hoc\Thong_Ke_May_Tinh\bai tap\datasets\bmi.csv", delimiter=',',dtype=str)
data3 = data3[1:]
print("Tong so nam:", np.sum(data3[:,1] == 'Male'))
print("Tong so nu:", np.sum(data3[:,1] == 'Female'))
# Ve bieu do hist
# Hist hien thi gia tri chieu cao va can nang cua mot nguoi
plt.subplot(2, 1, 1)
plt.hist(data3[:,2], bins=60)
plt.title("Chieu cao")
plt.subplot(2, 1, 2)
plt.hist(data3[:,3], bins=60)
plt.title("Can nang")
plt.show()
```

```python
# Su dung boxplot de so sanh chieu cao va can nang cua nam va nu
male_height = data3[data3[:, 1] == 'Male', 2].astype(float)
female_height = data3[data3[:, 1] == 'Female', 2].astype(float)
male_weight = data3[data3[:, 1] == 'Male', 3].astype(float)
female_weight = data3[data3[:, 1] == 'Female', 3].astype(float)

plt.figure(figsize=(12, 6))

# Boxplot for height
plt.subplot(1, 2, 1)
plt.boxplot([male_height, female_height], labels=['Male', 'Female'])
plt.title('Comparison of Heights')
plt.ylabel('Height')

# Boxplot for weight
plt.subplot(1, 2, 2)
plt.boxplot([male_weight, female_weight], labels=['Male', 'Female'])
plt.title('Comparison of Weights')
plt.ylabel('Weight')

plt.show()
```
