```python
from scipy.stats import poisson
from scipy.stats import norm
from scipy.stats import binom
import numpy as np
import matplotlib.pyplot as plt
import math
# Tham số của phân phối Poisson
lam = 5  # lambda = 5

# Xác suất P(X = 1)
p_x_1 = poisson.pmf(1, lam)

# Xác suất P(X <= 1)
p_x_leq_1 = poisson.cdf(1, lam)

# Xác suất có điều kiện P(X >= 1 | X <= 1)
p_ketqua = p_x_1 / p_x_leq_1

print(f"Xác suất có điều kiện P(X >= 1 | X <= 1) = {p_ketqua:.4f}")
```

```python
from scipy.stats import poisson

# Tham số Poisson
mu = 2.8

# Xác suất X = k với k = 0, 1, 2, 3
p_0 = poisson.pmf(0, mu)
p_1 = poisson.pmf(1, mu)
p_2 = poisson.pmf(2, mu)
p_3 = poisson.pmf(3, mu)

# Xác suất X >= 3
p_geq_3 = 1 - (p_0 + p_1 + p_2)

# Kỳ vọng E[Y]
E_Y = 0 * p_0 + 1 * p_1 + 2 * p_2 + 3 * (p_3 + p_geq_3)

# Số tiền trung bình thu được
revenue = 20 * E_Y - 24
revenue
```
