```python
df2 = pd.read_csv("../dataset/bds.csv")
df2.head(10)
```

```python
print(df2['Ngành cấp 3'].value_counts())
```

```python
print(df2.groupby("Sàn giao dịch")['Ngành cấp 3'].value_counts())
```

```python
# Vẽ từng biểu đồ tròn cho từng ngành cấp 3
for industry in df2["Ngành cấp 3"].unique():
    industry_data = df2[df2["Ngành cấp 3"] == industry]["Sàn giao dịch"].value_counts()

    plt.figure(figsize=(6, 6))
    plt.pie(industry_data, labels=industry_data.index, autopct='%1.1f%%', startangle=90)
    plt.title(f"Tỷ lệ sàn giao dịch - {industry}")
    plt.ylabel("")
    plt.show()
```

```python
hnx, hose = df2.groupby("Sàn giao dịch")['Ngành cấp 3'].value_counts().groupby(level=0).apply(list)
print(hnx)  # [11, 11]
print(hose) # [35, 23]
plt.subplot(1,2, 1)
plt.pie(hnx, labels=['BDS', 'PT'], autopct='%1.1f%%', startangle=90)
plt.subplot(1,2,2 )
plt.pie(hose, labels=['BDS', 'PT'], autopct='%1.1f%%', startangle=90)
plt.show()
```

```python
df3 = pd.read_csv("../dataset/Tweets.csv")
```

```python
df3.head(10)
```

```python
print("So luong phan hoi tung muc dco", df3['airline'].value_counts())
print("SO luong phan hoi", df3.groupby('airline')['airline_sentiment'].value_counts())
```

```python
# Với mỗi mức độ phản hồi của từng hãng máy bay, hãy xác định các tham số thống
# kê: giá trị trung bình, độ lệch chuẩn của các thuộc tính đô tin cậy của phản hồi
# (airline_sentiment_confidence), độ tin cậy của lý do phủ định
# (negativereason_confidence); và độ đo xu hướng tập trung (central tendency) của
# thuộc tính múi giờ (user_timezone).
print(df3.groupby(['airline', 'airline_sentiment'])[['airline_sentiment_confidence', 'negativereason_confidence']].agg(['mean', 'std']))


```

```python
print(df3.groupby(['airline','airline_sentiment'])['user_timezone'].describe())
```

```python

soLuongPhanHoi  = np.array(df3['airline'].value_counts())
print(soLuongPhanHoi)
plt.pie(soLuongPhanHoi, labels=df3['airline'].value_counts().index, autopct='%1.1f%%')
plt.title("So luong phan hoi cua moi hang may aby")
plt.show()
```

```python
sns.boxplot(data=df3, x= 'airline_sentiment_confidence', y= 'airline')
plt.title("Phan bo do tin cay")
plt.show()
```

```python
# Bằng đồ thị hộp râu (box plot) biểu diễn trực quan về cách dữ liệu của thuộc tính
# đô tin cậy của phản hồi (airline_sentiment_confidence) theo hãng hàng không nằm
# rải rác trên mặt phẳng, so sánh giữa các mức độ phản hồi.
plt.figure(figsize=(10, 6))
sns.boxplot(data=df3, x='airline', y='airline_sentiment_confidence', hue='airline_sentiment')
plt.title("Phân bố độ tin cậy của phản hồi theo hãng hàng không")
plt.xlabel("Hãng hàng không")
plt.ylabel("Độ tin cậy của phản hồi")
plt.xticks(rotation=45)
plt.show()
# 
```

```python
# Lọc các cột số để tính toán hệ số tương quan
numeric_df = df3.select_dtypes(include=['number'])

# Tính ma trận tương quan
correlation_matrix = numeric_df.corr()

# Vẽ heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)
plt.title("Biểu đồ Heatmap thể hiện tương quan giữa các thuộc tính số")
plt.show()
```
